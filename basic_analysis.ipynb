```python
import pandas as pd
import json
df = pd.read_csv('deardata.csv')
```

```python
def parse_text(row):
    t = json.loads(row[5])
    s = ''
    for item in t:
        s+=item['content']
    return s

df['doc'] = df.apply(parse_text, axis=1)
df.head()
```

```python
# word frequency
from textblob import TextBlob
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words("english"))
stop_words = stop_words.union(["’", "‘", "“"])
```

```python
# create a word set without stop words
wset = gset = sset = set()
wdoc = gdoc = sdoc = ''

for index, row in df.iterrows():
    db = TextBlob(row['doc'])
    ws = set(db.words)
    wdoc+=row['doc']
    wset = wset.union(ws)
    if row['author']=='Giorgia':
        gset = gset.union(ws)
        gdoc += row['doc']
    else:
        sset = sset.union(ws)   
        sdoc += row['doc']
        
wset = wset-stop_words
gset = gset-stop_words
sset = sset-stop_words

wb = TextBlob(wdoc)
wdic = {}
for word in wset:    
    wdic[word]=wb.word_counts[word]

gb = TextBlob(gdoc)
gdic = {}
for word in gset:    
    gdic[word]=gb.word_counts[word]

sb = TextBlob(sdoc)
sdic = {}
for word in sset:    
    sdic[word]=sb.word_counts[word]
    
    
wdf = pd.DataFrame.from_dict({'word':list(wdic.keys()), 'count':list(wdic.values())})
                              
wdf.sort_values('count', ascending=False)

# love, friends, drawing, data, 
```

```python
gdf = pd.DataFrame.from_dict({'word':list(gdic.keys()), 'count':list(gdic.values())})
                              
gdf.sort_values('count', ascending=False)

# people, boyfriend, friends, drawing, data, week, time, us, work
```

```python
sdf = pd.DataFrame.from_dict({'word':list(sdic.keys()), 'count':list(sdic.values())})
                              
sdf.sort_values('count', ascending=False)
# people, husband, life, drawing, data, week, time
```
